```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
pos_test_df = pd.read_csv("Test_on_day_of_result.csv", sep=" ", header=None, index_col=0, parse_dates=True)
#pos_test_df.drop(index=["I", "Antal"], inplace=True)
pos_test_df = pos_test_df[~ pos_test_df.index.isnull()]
pos_test_df.rename(columns={2:"Svardato", 3:"Pr√∏vedato"}, inplace=True)
week_indicator = ["Ma", "Mandag"]
indicators = pos_test_df[pos_test_df[1]==week_indicator[0]]
pos_test_df.index
```

```python
pos_test_df.plot(figsize = (15,5)) 
plt.xlabel("Dato")
plt.ylabel("Positive tests")
for i in indicators.index:
    # print(i)
    plt.axvline(i, color="green", label=week_indicator[1])
plt.legend(["Dato", "Positive tests", "Mandag"])
plt.savefig("proevesvar.pdf")
```
